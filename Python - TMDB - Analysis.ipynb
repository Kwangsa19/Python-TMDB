### Use weighted average formula from IMDB to find the weighted rating for each movie (Fair calculation)

The following formula is used to calculate the Top Rated 250 titles. This formula provides a true 'Bayesian estimate', which takes into account the number of votes each title has received, minimum votes required to be on the list, and the mean vote for all titles:

weighted rating (WR) = (v  (v+m))  R + (m  (v+m))  C

Where:

R = average for the movie (mean) = (rating)

v = number of votes for the movie = (votes)

m = minimum votes required to be listed in the Top Rated 250 list (currently 25,000)

C = the mean vote across the whole report

Please visit this [link](https://help.imdb.com/article/imdb/track-movies-tv/ratings-faq/G67Y87TFYYP6TWAV#)

```python
# C = the mean vote acroos the whole report
C = df['vote_average'].mean()
print(C)
```

```python
# Minimum number of votes required to be in the chart 
# we could take 80th percentile or 85th percentile. In this case, I just chose 80th percentile. 
m = df['vote_count'].quantile(0.80)
print(m)
```

```python
def rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
#applying this to our new dataset:
df['w_rating'] = df.apply(rating, axis=1)
```

```python
df.head(200)
```

```python
# Plotting graph for the top 10 movies which has the most weighted average (Sorted)

# Sort the DataFrame by vote_count in descending order
df = df.sort_values(by='w_rating', ascending=False)

# Get the top 10 titles 
top_10_titles = df['title'].head(10)
top_10_vote_counts = df['w_rating'].head(10)

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.barh(top_10_titles, top_10_vote_counts, color='skyblue')
plt.xlabel('Weighted Rating')
plt.ylabel('Title')
plt.title('Top 10 Titles by Weighted Average')
plt.gca().invert_yaxis()  # Invert the y-axis to display the highest weighted rating at the top

# Show the plot
plt.tight_layout()
plt.show()
```
